### Design a multi-line plot using Matplotlib or Seaborn that shows the evolution of average mpg over the years, separately for each origin

```python
#your code here
# Aggregate mean mpg by year and origin
avg_mpg = df_cars.groupby(['model_year', 'origin'])['mpg'].mean().reset_index()

plt.figure(figsize=(10,6))

# Get unique origins
origins = avg_mpg['origin'].unique()

for origin in origins:
    data = avg_mpg[avg_mpg['origin'] == origin]
    plt.plot(data['model_year'], data['mpg'], marker='o', label=origin)

plt.title('Average MPG over Years by Origin')
plt.xlabel('Year')
plt.ylabel('Average MPG')
plt.grid(True)
plt.legend(title='Origin')
plt.show()
```

### Create a Seaborn scatterplot (or PairGrid) where:

X = horsepower

Color by: origin

Size by: mpg

Hue order = ['japan', 'europe', 'usa']

Add meaningful plot titles and axis titles.

```python
#your code here
df=pd.read_csv('Cars.csv')
plt.figure(figsize=(10, 7))

sns.scatterplot(
    data=df,
    x='horsepower',
    y='weight',
    hue='origin',
    size='mpg',
    hue_order=['japan', 'europe', 'usa'],
    palette='Set1',  # You can choose any palette you like
    sizes=(20, 200),  # Control min and max marker sizes
    alpha=0.7
)

plt.title('Car Weight vs Horsepower Colored by Origin and Sized by MPG')
plt.xlabel('Horsepower')
plt.ylabel('Weight')
plt.legend(title='Origin', bbox_to_anchor=(1.05, 1), loc='upper left')  # Places legend outside plot

plt.tight_layout()
plt.show()
```

### We define a consistent car model as one that was produced over multiple years and had very low variation in mpg across those years (standard deviation < 1.0).

Tasks:

Identify car names that appear in more than one model_year.

For each such name, compute the standard deviation of mpg across years.

Return the car(s) with the lowest variation in mpg, among those with at least 2 appearances and std(mpg) < 1.0.

Report the model name(s), number of appearances, and the average mpg.

Bonus:
Sort the result by number of appearances (descending), then mpg (descending).

```python
#your code here

name_year_counts = df.groupby('name')['model_year'].nunique()
multi_year_names = name_year_counts[name_year_counts > 1].index

multi_year_cars = df[df['name'].isin(multi_year_names)]

stats = multi_year_cars.groupby('name').agg(
    mpg_std=('mpg', 'std'),
    appearances=('mpg', 'count'),
    avg_mpg=('mpg', 'mean')
).reset_index()

consistent_cars = stats[(stats['mpg_std'] < 1.0) & (stats['appearances'] >= 2)]

min_std = consistent_cars['mpg_std'].min()
lowest_variation_cars = consistent_cars[consistent_cars['mpg_std'] == min_std]

result = lowest_variation_cars.sort_values(by=['appearances', 'avg_mpg'], ascending=[False, False])

# Display result
print(result[['name', 'appearances', 'avg_mpg']])
```
